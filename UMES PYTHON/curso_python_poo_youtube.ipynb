```python
#Herencia en POO 

class Persona:
 def __init__(self, nombre, edad, nacionalidad):
  self.nombre = nombre 
  self.edad = edad 
  self.nacionalidad = nacionalidad

 def hablar(self):
  print("Hola, Estoy Hablando...")

class Empleados(Persona):
    def __init__(self, nombre, edad, nacionalidad, trabajo, salario): #aqui es necesario colocar todos los atributos de la clase anterior 
        super().__init__(nombre, edad, nacionalidad) #aqui solo llamamos a los atributos que necesitamos
        self.trabajo = trabajo
        self.salario = salario

roberto = Empleados("Roberto", 34, "guatemalteco", "Programador", 1000000)



```

```python
#Vamooos a ver Herencia Multiples


class Persona:
 def __init__(self, nombre, edad, nacionalidad):
  self.nombre = nombre 
  self.edad = edad 
  self.nacionalidad = nacionalidad

 def hablar(self):
  print("Hola, Estoy Hablando...")

class Artista:
 def __init__(self, habilidad):
  self.habilidad = habilidad

 def mostrar_habilidad(self):
  print(f"mi habilidad es: {self.habilidad}")

#QUE PASARIA SI YO QUIERO HACER UNA CLASE EMPLEADO ARTISTA 

class EmpleadoArtista(Persona, Artista): #TIENE DOS CLASES que van a herederar esos atributos. 
 def __init__(self, nombre, edad, nacionalidad, habilidad, salario, empresa):
  Persona.__init__(self, nombre, edad, nacionalidad ) #ahora ya no ponemos super() si no directamente el nombre de la clase
  Artista.__init__(self, habilidad)
  self.salario = salario  #Creamos los atributos faltantes 
  self.empresa = empresa 

 def presentarse(self):
  return f"Hola, soy: {self.nombre} y Mi habilidad es {self.habilidad} y trabajo en {self.empresa} " #super() le dara a entender a python que estamos heredando ese metodo, accediendo a la clase padre, es decir la clase base 

artista1 = EmpleadoArtista("Jose", 19, "guatemalteca", "Programar", 1000000, "Amazon")

artista1.presentarse()

```

```python
#EJERCICIO MIO  No 1 

class Motocicleta():
  def __init__(self, color, matricula, combustible_litros, numero_ruedas, marca, modelo, fecha_fabricacion, velocidad_punta, peso):
    self.color = color 
    self.matricula = matricula
    self.combustible_litros = combustible_litros
    self.numero_ruedas = numero_ruedas
    self.marca = marca 
    self.modelo = modelo
    self.fecha_fabricacion = fecha_fabricacion
    self.velocidad_punta = velocidad_punta
    self.peso = peso 
    self.motor = True 

  def metodo_arrancar(self):
   if self.combustible_litros <= 0:
    print(f"No haya suficiente {self.combustible_litros}")
   elif self.motor == True: 
    print(f"El motor esta en marcha {self.motor}")
   else: 
    print(f"Se esta preparando para arrancar.")
    self.motor = True 
  
  def metodo_detener(self):
   if self.motor == True:
    print("El motor se esta deteniendo...")
    self.motor = False
   else: 
    print("Ya el motor estaba apagado.")

moto1 = Motocicleta("verde", "Mxz39", 50, 4, "Toyota", "2024", "2000", 250, 100)

print(moto1.metodo_arrancar())
print(moto1.metodo_detener())



```
